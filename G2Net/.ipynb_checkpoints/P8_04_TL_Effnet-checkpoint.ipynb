```python
DATA_PATH = '../data/G2Net/g2net-gravitational-wave-detection/'

TB_LOG_PATH = '../data/G2Net/tb_logs'
AUTOTUNE = tf.data.AUTOTUNE

SPLIT_RATIO = .9
RANDOM_STATE = 42

IMG_WIDTH = 69
IMG_HEIGHT = 193
```

```python
train = tf.data.experimental.load(os.path.join(DATA_PATH, 'ds_train'))
valid = tf.data.experimental.load(os.path.join(DATA_PATH, 'ds_valid'))
```

```python
def prep(img, label):
    ret = preprocess_input(img)
    return ret,label
```

```python
BATCH_SIZE = 16

# train_data = train.take(250000).map(prep).batch(BATCH_SIZE).prefetch(buffer_size=64)
# valid_data = valid.take(25000).map(prep).cache().batch(BATCH_SIZE).prefetch(buffer_size=64)
train_data = train.take(200000).batch(BATCH_SIZE).prefetch(buffer_size=64)
valid_data = valid.take(20000).cache().batch(BATCH_SIZE).prefetch(buffer_size=64)
```

```python
print(train_data.cardinality)
```

```python
from tensorflow.keras.applications.efficientnet import *
from tensorflow.keras.applications.efficientnet import preprocess_input

base_model = EfficientNetB0(
    include_top=False,
    input_shape=(IMG_WIDTH, IMG_HEIGHT,3),
    weights='imagenet',
    drop_connect_rate=0.3),

# base_model = ResNet50V2(
#     weights='imagenet',
# #     weights=None,
#     include_top=False,
#     input_shape=(IMG_WIDTH, IMG_HEIGHT,3),
#     pooling='avg')
# base_model.trainable=True
# base_model.trainable=False


fine_tune_at = 171
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable =  False
```

```python
metrics = [tf.keras.metrics.AUC(name='AUC'),
        tf.keras.metrics.BinaryAccuracy(name='Accuracy'),]

lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-4,
    decay_steps=5000,
    decay_rate=0.9)

tb_callback = tf.keras.callbacks.TensorBoard(
    os.path.join(TB_LOG_PATH,'Res2_base_1e4'),
    update_freq=64,
    )



callbacks= [
    tb_callback,
]

def create_model(alpha):
    model = tf.keras.Sequential([
        Layer.InputLayer(input_shape=(IMG_WIDTH, IMG_HEIGHT,1)),
#         Layer.experimental.preprocessing.Rescaling(255),
        Layer.Conv2D(3,3,activation='relu',padding='same'),
#         Layer.Lambda(preprocess_input),
#         Layer.Dropout(.3),
        base_model,
        Layers.GlobalAveragePooling2D(name="avg_pool")
        Layers.BatchNormalization()
        Layers.Dropout(.2, name="top_dropout")
        #         Layer.Dropout(.3),
#         Layer.Dense(32,activation='relu'),
        Layer.Dense(1, activation='sigmoid')])

#     Layer.Dense(256),
#         Layer.LeakyReLU(alpha=alpha),
#         Layer.Dense(256),
#         Layer.LeakyReLU(alpha=alpha),
# #         Layer.Dense(8),
# #         Layer.Dropout(.3),
# #         Layer.LeakyReLU(alpha=alpha),
#         Layer.Dense(1, activation='sigmoid'),
#     ])
    return model
```

```python
model = create_model(.3)
model.summary()
```

```python
print("Number of layers in the base model: ", len(base_model.layers))
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(
    learning_rate=lr_schedule,
    ),
    loss=tf.keras.losses.BinaryCrossentropy(), 
    metrics=metrics)
```

```python
model.fit(
    train_data,
    epochs=2,
    validation_data=valid_data,
    callbacks = callbacks,
    )
```

```python
pred = model.predict(valid_data.take(100))
```
